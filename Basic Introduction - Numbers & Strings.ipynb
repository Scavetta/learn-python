#### Numbers

   - Run the below code snippets either in Command Prompt/Terminal Interactive Session by running the command **python** or just put all the snippets in the single python file. To know how to create a python file, please refer the previous section of this tutorial
   ```
    >>> 10 + 20       # 30 (Plus - Addition)
    >>> 20 - 10       # 10 (Dash/Hyphen - Subtraction)
    >>> 2 * 5         # 10 (Asterisk - Multiplication)
    >>> 5 / 3         # 1.6666666666666667 (Slash - Division with decimals)
    >>> 5 // 3        # 1 (Double Slash - Quotient value of a division)
    >>> 5 % 3         # 2 (Modulo Operator - Remainder value of a division)
    >>> 3 ** 10       # 59049 (3 to the power of 10)
    >>> a = 20
    >>> b = 2
    >>> a * b + 20 + 0.5    # 60.5
   ```
   - Please refer the operator precedence to know how the above expression is evaluated
[Click here for Operator Precedence](https://docs.python.org/3/reference/expressions.html#operator-precedence)
   - If you want to try the above code snippets in a python file, don't include the >>> symbol in a python file
   
#### Strings
   - Python can also work with Strings, not only with Numbers
   - String in a Python can be enclosed with ```'...'``` or ```"..."```
   ```
    >>> 'Hello there'           # Line 1
    >>> "Hello there"           # Line 2
    >>> 'Don\'t go anywhere'    # Line 3
    >>> "Don't go anywhere"     # Line 4
    >>> "\name"                 # Line 5
    # >>> print("\name")          # Line 6
    >>> """
            You can 
            also 
            include multiple lines like this
        """                     # Line 7
    >>> 10 * 'ab'               # Line 8
    >>> 10 * 'ab' + 'cd'        # Line 9
    >>> 'hi' 'bye'              # Line 10
    >>> a = 'hi'                # Line 11
    >>> b = 'bye'               # Line 12
    >>> a + b                   # Line 13
    >>> a b                     # Line 14  (Will throw an error)
    >>> print('to '
            'break '
             'long '
             'text.')
   ```
   - Advanced methods to play with Characters in String
   ```
    >>> word = 'Righteousness'
    >>> word[0]             # R
    >>> word[4]             # t
    >>> word[:]             # Righteousness
    >>> word[4:]            # teousness
    >>> word[4:8]           # teou
    >>> word[4:100]         # teousness
    # >>> word[15]            # *Will throw an error*
    >>> word[:5]            # Right
    >>> word[:5] + word[5:9] # Righteous
    >>> word[-1]            # s (last character)
    >>> word[-3]            # e (from last character)
    >>> word[-3:]           # ess
    >>> word[:-4]           # Righteous
    >>> word[0:9:2]         # R*g*t*o*s => Rgtos
    >>> len(word)           # 13
   ```

### Input from User

- ```input()``` is a function in Python which helps to get an input from user at runtime
- ```input()``` always returns the value in a String (str) datatype
```
    a = input()        # Getting a user input, value should be 10
    print(a)           # Printing a value
    print(type(a))     # Print its datatype
    b = int(a)         # Parse into Integer datatype
    print(type(b))
```
