### Control Statements

Control Statements are mostly common in all the programming languages to decide what to do, how to do and number of times should do

#### IF Statement
```
    if <expression>:
        <statements>
    elif <expression>:
        <statements>
    else:
        <statements>
```

You can add **multiple elif** (else if) in a nested way or just go with **simple if** (only if, without elif and else) or **can use if along with else** (without elif)

#### Break

**break** statement in a python can be used to terminate a **loop** without executing all the iterations when satisfying certain condition(s)

```
    for i in range(1, 5):
        if i % 2 == 0:
            break;
        print(i)
```

#### Continue
**continue** statement in a python can be used to go to next iteration in a **loop** when certain condition(s) satisfied without executing a further statements after the **continue** statement

```
    for i in range(1, 5):
        if i % 2 == 0:
            continue;
        print(i)
```

#### Pass
**pass** statement in a python used to when a statement is required syntactically, but the no action required by the program
```
    for i in range(0,5):
        pass
```

### Function

Functions are the primary component in all the programming languages which is mainly used to divide the lenthier code into a small portions

```
   def <function_name>(parameter_1, parameter_2, [...]):
       <statements>
       [return]
```

In Python, no need to define any return type of the function and even **return** statement also an optional

```
    def list_numbers(n):
        a = []
        for i in range(n):
            a.append(i)
        print(a)
        return a
        
    list_numbers(6)                 # Comment the return statement in list_numbers()
    print(list_numbers(6))          # Comment the print statement in list_numbers()
```

#### lambda function

**lambda function** is an anonymous function which can expressed in a single line of statements

```
a = lambda x: x + 2
print(a(2))
print([a(i) for i in range(5)])
```
